### Problem Data

```python
# numerical parameters
nel = int(1000) # number of elements
tf = 15.0 # total calculation tim (sec)

#problem Data
Tc = 10. # time for total valve closure (sec)
degTheta = 90.0 # 90->vertical well, 0->horizontal well
p0 = 20 # reservoir pressure (MPa)
q0 = 0.2*3600 # SS flow rate (m3/hr)

mu = np.array([1.60 * 10**-3]) # Dynamic viscosity (Pa-s, N-s/m2)
sg = np.array([1.0]) # specific gravity
rough = np.array([0.9]) # roughness (mm)
d = np.array([273.1]) # mm
a = np.array([1000.]) # speed of sound in fluid (m/s)
L = np.array([1000.]) # length of each conduit reach (m)
```

### Check Fluid Velocity

```python
area = np.pi * (d[0]/1000)**2 / 4. #m 2
velocity = q0 / area / 3600.
print(velocity) # m/sec
```

### Calculated Constants

```python
nnode = nel + 1 # number of nodes
x = np.linspace(0., sum(L), nnode) # spatial discretization
nrlp = np.sum(x[:]>=L[-1])# number of reachs in last segment

dx = sum(L)/nel
s = 0.03125*dx/max(a) # time step size (s)
t = np.linspace(0, tf, tf/s + 1) # time discretization
```

### Courant Evaluation

```python
Courant = max(a)*s/dx; Courant
print(Courant)
print(s)
print(dx)
```

### Fluid / Conduit specific values

```python
A = np.pi * (d / 1000.)**2 / 4. # area of flow (m2)
ani = L / (s*a)
g = 9.806 # m/s2
rho = sg * 1000 # kg/m3

f = np.zeros(len(d))
ca = np.zeros(len(d))
cf = np.zeros(len(d))
cb = np.zeros(len(d))
for i in xrange(len(d)):
    f[i] = sf.SS_DeltaH(q0, mu[i], sg[i], d[i], rough[i], x, degTheta, units=1).Calc().f # friction factor 
    ca[i] = g * A[i] / a[i] # m2/sec
    cf[i] = s * f[i]/(2*(d[i]/1000.)*A[i]) #assumed constant (sec/m3)
    cb[i] = s * g * A[i] * np.sin(degTheta*np.pi / 180.0)
```

```python
print(f) #friction factor
```

### Create idx Array - Determines where changes in system occur (e.g., conduit diameter, fluid, etc.)

```python
idx = np.zeros(len(x)).astype(int) # array of integers
chng = np.arange(0,len(d),1) # each index of the 'chng' array indicates a change in conduit diameter
for i in xrange(len(idx)):
    idx[i] = chng[0]
```

### Steady State Calc

```python
h0_arr = np.zeros(len(x)) # head (m)
q0_arr = np.zeros(len(x)) # flow rate (m3/hr)
p0_arr = np.zeros(len(x)) # pressure (MPa)
    
for i in xrange(len(x)):
    q0_arr[i] = q0 / 3600. # steady flow (m3/sec)
    h0_arr[i] = (p0*10**6)/(rho[idx[i]]*g) + sf.SS_DeltaH(q0, mu[idx[i]], sg[idx[i]], d[idx[i]], rough[idx[i]],
                                                  x[i], degTheta, units=1).Calc().dh # head (m)
    p0_arr[i] = p0+sf.SS_DeltaH(q0,mu[idx[i]],sg[idx[i]],d[idx[i]],rough[idx[i]],x[i],degTheta,units=1).Calc().dp
```

### Main Program

Compile main loop in C using Numba's Just In Time compiler (JIT)

```python
from numba import jit
```

build arrays outside of the main function

```python
# rows-> spatial disc; column -> temporal disc
row = x.size
col = t.size
qp = np.empty(row, dtype = np.float) # future flow rate along the positive characteristic (m3/hr)
hp = np.empty(row, dtype = np.float) # future head along the positive characteristic (m3/hr)

hin = np.empty((row, col), dtype = np.float) # head (m)
qin = np.empty((row, col), dtype = np.float) # flow rate (m3/hr)
pin = np.empty((row, col), dtype = np.float) # pressure (MPa)
```

the compiled function

```python
@jit
def main2(x, t, q0_arr, h0_arr, p0_arr, ca, cf, cb, idx,
         row, col, qp, hp, h, q, p):
    
    # define initial conditions
    for i in range(row):
        q[i,0] = q0_arr[i]
        h[i,0] = h0_arr[i]
        p[i,0] = p0_arr[i]
    
    # initial values at upstream reservoir
    cn = q[1,0] - ca[0]*h[1,0] - cb[0] - cf[0] * q[1,0] * np.abs(q[1,0]) # negative characteristic (m3)
    qp[0] = cn + ca[0]*h[0,0]
    
    # interior nodes
    for j in range(col - 1):
        for i in range(row - 2):
            i = i + 1 # skip first value (BCT)
            ip1 = i + 1
            im1 = i - 1
            cn = q[ip1, j] - ca[idx[i]]*h[ip1,j] - cb[idx[i]] - cf[idx[i]]*q[ip1,j]*np.abs(q[ip1,j])
            cp = q[im1, j] + ca[idx[i]]*h[im1,j] - cb[idx[i]] - cf[idx[i]]*q[im1,j]*np.abs(q[im1,j])
            qp[i] = 0.5 * (cn + cp)
            hp[i] = (cp - qp[i])/ ca[idx[i]]
            if i == 1: # calculations for upstream reservoir BCTs
                hp[0] = h[0,0] # Constant upstream reservoir head (m)
                qp[0] = cn + ca[idx[i]]*hp[0]
            if i == row - 2: # calculations for downstream valve BCTs
                #interpolate valve 
                interp = t[j]
                if interp >= Tc:
                    tau = 0
                else:
                    I = 0
                    while interp >= tc[I]:
                        I = I + 1
                    frac = (interp - tc[I-1]) / (tc[I] - tc[I-1])
                    tau = y[I-1] + frac * (y[I] - y[I-1])
        
                cv = (tau * q[row-1,0] * q[row-1,0]) / (ca[idx[i]] * h[row-1,0])
                qp[i+1] = 0.5 * (-cv + ((cv*cv) + 4 * cp * cv)**(0.5))
                hp[i+1] = (cp - qp[i])/ ca[idx[i]]
        q[:,j+1] = qp
        h[:,j+1] = hp
    return q, h
```
